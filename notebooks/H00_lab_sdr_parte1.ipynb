## Tutorial de Instalação Configuraçã GNU Radio 3.9 no UBUNTU 20.04
## Objetivos
As metas desse tutorial são ajudar o usuário a:
- Instalar e configurar uma maquina virtual Linux Ubuntu 20.04 LTS (Focal Fossa).


**Passo 1**: Baixe e instale o VirtualBox em seu computador (para acessar a página de Download [acesse aqui](https://www.virtualbox.org/)).

![Virtual1](./DCO2010_2022/FIGS/H00/virtualbox/virtual_box.png)

**Passo 2**: Baixe a iso do Ubuntu 20.04-3 LTS (Focal Fossa).
- Versão 64 Bits https://releases.ubuntu.com/20.04/ubuntu-20.04.3-desktop-amd64.iso

**Passo 3**: Após instalar o Virtualbox e baixar a iso, abra o VirtualBox e clique em **Novo **, e depois na opção **Criar** de acordo com a imagem :
![fig_Virtual2](./DCO2010_2022/FIGS/H00/virtualbox/virtual_box1.png)


**Passo 4**: Escolha a memória que deseja alocar, o ideal é que seja a metade da memória de sua maquina :
![fig_Virtual2](./DCO2010_2022/FIGS/H00/virtualbox/virtual-box-memoria.png)


**Passo 5**: Nessa tela escolha a opção ** VDI (VirtualBox Disk Image)** e prossiga a instalção:
![fig_VM2](./DCO2010_2022/FIGS/H00/virtualbox/virtual_box2.png)

**Passo 6**: Escolha a opção ** Dinamicamente Alocado **:
![fig_VM3](./DCO2010_2022/FIGS/H00/virtualbox/virtual_box3.png)

**Passo 7**: Informe o nome do disco virtual e especifique **60 GB** de espaço em HD (se tiver pouco espaço em HD, pode criar uma máquina de **40 GB**): :
![fig_VM3](./DCO2010_2022/FIGS/H00/virtualbox/virtual_box4.png)

**Passo 8**: Inicialize a máquina criada com dois cliques sobre ela:
![fig_VM3](./DCO2010_2022/FIGS/H00/virtualbox/virtual_box5.png)

**Passo 9**: Localize a iso do ubuntu baixada no ícone da pasta:
![fig_VM3](./DCO2010_2022/FIGS/H00/virtualbox/virtual_box6.png)

**Passo 10**: Selecione a iso, para começar o processo de instalação:
![fig_VM3](./DCO2010_2022/FIGS/H00/virtualbox/virtual_box7.png)

**Passo 11**: Espere carregar a máquina e continuar o processo de instalação:
![fig_VM3](./DCO2010_2022/FIGS/H00/virtualbox/ubuntu1)

**Passo 12**: Aguarde enquanto carrega :
![fig_VM3](./DCO2010_2022/FIGS/H00/virtualbox/inicio_ubuntu)

**Passo 13**: Escolha o idoma para prosseguir selecione a opção Install para prosseguir a instalação:
![fig_VM3](./DCO2010_2022/FIGS/H00/virtualbox/ubuntu2)

**Passo 14**:Escolha o idioma do teclado:
![fig_VM3](./DCO2010_2022/FIGS/H00/virtualbox/ubuntu4)

**Passo 15**: Nas opções abaixo configure de acordo com a imagem:
![fig_VM3](./DCO2010_2022/FIGS/H00/virtualbox/ubuntu6)

**Passo 16**: escolha a primeira opção e continue a instalação:
![fig_VM3](./DCO2010_2022/FIGS/H00/virtualbox/ubuntu7)

**Passo 17**: Nessa tela click na opção **continue**:
![fig_VM3](./DCO2010_2022/FIGS/H00/virtualbox/ubuntu8)

**Passo 18**: Informe sua localização e click em **continue** novamente:
![fig_VM3](./DCO2010_2022/FIGS/H00/virtualbox/ubuntu9)

**Passo 19**: Preencha as informações,referentes a máquina como desejar.:
![fig_VM3](./DCO2010_2022/FIGS/H00/virtualbox/ubuntu10)

**Passo 20**: Aguarde enquanto é feita instalção :
![fig_VM3](./DCO2010_2022/FIGS/H00/virtualbox/ubuntu11)

**Passo 21**: Espere carregar a máquina e continuar o processo de instalação: 
![fig_VM3](./DCO2010_2022/FIGS/H00/virtualbox/ubuntu12)

**Passo 22**: Assim que concluída escolha a opção **reset**:
![fig_VM3](./DCO2010_2022/FIGS/H00/virtualbox/ubuntu14)

**Passo 23**: Sua máquina virtual está pronta para uso:
![fig_VM3](./DCO2010_2022/FIGS/H00/virtualbox/ubuntu15)
