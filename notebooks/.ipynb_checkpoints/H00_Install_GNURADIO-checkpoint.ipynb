## GNU Radio 3.9 instalação 
## Objetivo
- Instalar e configurar o GNU Radio 3.9 no Ubuntu 20.04.3 LTS (Focal Fossa) com suporte a USRP.

## Parte 1: Instalação do UHD (Para uso do USRP)

De acordo com o site https://wiki.gnuradio.org/index.php/InstallingGR#From_Source para utilizar a USRP junto  com o GNU Radio é necessário instalar software USRP Hardware Driver (UHD). 

Segue abaixo os passos para instalação a partir do código fonte. Seguindo as instruções de  https://kb.ettus.com/Building_and_Installing_the_USRP_Open-Source_Toolchain_(UHD_and_GNU_Radio)_on_Linux .

Abra o terminal do linux, pode ser feito de duas formas: com o atalho **Ctrl+Alt+T** ou pesquisando na Home a palavra terminal:


**Passo 1:** Para dar inicio a instalação utilize o comando a seguir:

```bash
sudo apt-get update
```
Você verá uma saída similar a:

![sudoaptgetupdate](../FIGS/H00/uhd/sudo-apt-get_update.png)

**Passo 2:** Instalação de dependências necessárias:

```bash
sudo apt-get -y install autoconf automake build-essential ccache cmake cpufrequtils doxygen ethtool fort77 g++ gir1.2-gtk-3.0 git gobject-introspection gpsd gpsd-clients inetutils-tools libasound2-dev libboost-all-dev libcomedi-dev libcppunit-dev libfftw3-bin libfftw3-dev libfftw3-doc libfontconfig1-dev libgmp-dev libgps-dev libgsl-dev liblog4cpp5-dev libncurses5 libncurses5-dev libpulse-dev libqt5opengl5-dev libqwt-qt5-dev libsdl1.2-dev libtool libudev-dev libusb-1.0-0 libusb-1.0-0-dev libusb-dev libxi-dev libxrender-dev libzmq3-dev libzmq5 ncurses-bin python3-cheetah python3-click python3-click-plugins python3-click-threading python3-dev python3-docutils python3-gi python3-gi-cairo python3-gps python3-lxml python3-mako python3-numpy python3-numpy-dbg python3-opengl python3-pyqt5 python3-requests python3-scipy python3-setuptools python3-six python3-sphinx python3-yaml python3-zmq swig wget
```

**Passo 15:** 

```bash
sudo ldconfig
```

Você verá uma saída similar a:


![Virtual1](../FIGS/H00/uhd/sudo_ldconfig.png)


**Passo 16:** Por fim, certifique-se de que a variável de ambiente LD_LIBRARY_PATH esteja definida e inclua a pasta na qual o UHD foi instalado. Mais comumente, você pode adicionar a linha abaixo ao final do seu arquivo

```bash
gedit $HOME/.bashrc &
```

Adicione a seguinte linha no arquivo e salve:

```bash
export LD_LIBRARY_PATH=/usr/local/lib:$LD_LIBRARY_PATH
```

Para que essa alteração tenha efeito, você precisará fechar a janela do terminal atual e abrir um novo terminal.

**Passo 17:** Você pode testar isso rapidamente, sem nenhum dispositivo USRP conectado, executando no terminal:

```bash
uhd_find_devices
```

### Testes e Problemas Clássicos de instalação

1- O GNURadio pode ser aberto pelo terminal com o comando:

```bash
gnuradio-companion
```

Você poderá ver uma saída similar a:

![Virtual1](../FIGS/H00/gnuradio/terminal.png)

2- Ele também pode ser aberto pela sua área de trabalho:

![Virtual1](../FIGS/H00/gnuradio/gnuradio-home.png)

3- Um problema recorrente na instalação pode ser visto na imagem a seguir, informações sobre a correção pode ser encontrada: https://wiki.gnuradio.org/index.php/ModuleNotFoundError .

![Virtual1](../FIGS/H00/gnuradio/import_gnuradio.png)

Para solução:

1. Digite: gnuradio-config-info --prefix
    Minha saída foi: /usr/local/.

2. Digite: find /usr/local/. -name gnuradio | grep "packages"
    Miinha saída foi: /usr/local/./lib/python3/dist-packages/gnuradio
    
3. Abra o seu arquivo de perfil, digitando: gedit $HOME/.bashrc &

e adicione a seguinte linha no arquivo e salve :

```bash
export PYTHONPATH=/usr/local/lib/python3/dist-packages:/usr/local/lib/python3/site-packages:$PYTHONPATH
```
Você poderá ver um arquivo similar a:

![Virtual1](../FIGS/H00/gnuradio/arquivo.png)

4- Se você obtiver o erro **ModuleNotFoundError** ao iniciar o GNU Radio a partir da GUI, mas puder iniciar o GNU Radio usando o terminal, é porque as variáveis de ambiente foram definidas para o terminal, mas não para seu ambiente gráfico. Experimente uma das soluções listadas a seguir:

1. Armazenar os comandos de exportação indicados em ~/.profile. Para abrir o arquivo de perfil, digite: gedit ~/.profile &

Adicione as seguintes linhas no arquivo e salve: 

```
export PYTHONPATH=/usr/local/lib/python3/dist-packages:/usr/local/lib/python3/site-packages:$PYTHONPATH
export LD_LIBRARY_PATH=/usr/local/lib:$LD_LIBRARY_PATH
```

Você poderá ver um arquivo similar a:

![Virtual1](../FIGS/H00/gnuradio/profile-linhas.png)

Reinicie o seu Linux.

2. Caso ainda não tenha dado certo, armazene os comandos de exportação indicados em um novo script de shell (com qualquer nome de sua escolha) em /etc/profile.d/. Reinicie o seu computador e tente novamente.
